- 도별 남,북 통합
- gangwon, gwangju는 test dataset 내 존재하지 않음 
  - gwangju는 jeolla로 대치

```python
def sido_merge(x):
    # 도별 남,북을 합침 
    if ((x=='choongnam')|(x=='choongbuk')):
        return 'choongcheong'
    elif ((x=='jeonnam')|(x=='jeonbuk')):
        return 'jeolla'
    elif ((x=='gyeongnam')|(x=='gyeongbuk')):
        return 'gyeongsang'

    # 광주 -> 전라
    elif (x=='gwangju'):
        return 'jeolla'

    # 대전, 세종 -> 충청
    elif ((x=='daejeon')|(x=='sejong')):
        return 'choongcheong'

    # 울산, 대구 -> 경상
    elif ((x=='daegu')|(x=='ulsan')):
        return 'gyeongsang'

    # 인천 -> 경기
    elif (x=='incheon'):
        return 'gyeonggi'
    
    else:
        return x
    
    
all_data['sido'] = all_data['sido'].apply(sido_merge)
```

```python
# 'inst_id','sgg' 칼럼 삭제
all_data.drop('inst_id', axis=1, inplace=True) #'inst_id': 고유 식별자로 불필요
all_data.drop('sgg', axis=1, inplace=True) #'sgg': 지역 정보는 'sido'열 존재하므로 삭제

# 'OC' 레이블값 대체
all_data['OC'].replace('open', 1, inplace=True)
all_data['OC'].replace(' close', 0, inplace=True)

# 'employee' 칼럼 float형으로 변환
all_data[len(train):]['employee1'] = all_data[len(train):]['employee1'].str.replace(',','')
all_data[len(train):]['employee2'] = all_data[len(train):]['employee2'].str.replace(',','')
all_data['employee1'] = all_data['employee1'].astype('float64')
all_data['employee2'] = all_data['employee2'].astype('float64')
```

```python
# 지역 정보 일반화하는 함수 생성
def sido_func(x):

    # 도별 남,북을 합침 
    if ((x=='choongnam')|(x=='choongbuk')):
        return 'choongcheong'
    elif ((x=='jeonnam')|(x=='jeonbuk')):
        return 'jeolla'
    elif ((x=='gyeongnam')|(x=='gyeongbuk')):
        return 'gyeongsang'

    # 광주 -> 전라
    elif (x=='gwangju'):
        return 'jeolla'

    # 대전, 세종 -> 충청
    elif ((x=='daejeon')|(x=='sejong')):
        return 'choongcheong'

    # 울산, 대구 -> 경상
    elif ((x=='daegu')|(x=='ulsan')):
        return 'gyeongsang'

    # 인천 -> 경기
    elif (x=='incheon'):
        return 'gyeonggi'
    
    else:
        return x

# 'sido' 칼럼 일반화
all_data['sido'] = all_data['sido'].apply(sido_func)
```

```python
# 'openDate' 칼럼 형 변환 및 Year 칼럼으로 추출
all_data['openDate'] = pd.to_datetime(all_data['openDate'].fillna(0).astype('int').astype('str'),
                                      format="%Y/%m/%d", errors='coerce')
all_data['openYear'] = all_data['openDate'].dt.year

# 'runYear' 칼럼 추가 생성, 'openYear','openDate' 칼럼 삭제
all_data['runYear'] = 2016 - all_data['openYear'] # 'runYear': 운영기간
all_data.drop(['openDate','openYear'], axis=1, inplace=True)
```

### ~~Yeo-Johnson Transformation~~
- 음수 값이 포함되어 있으므로 양수 값에만 적용 가능한 Box-Cox 변환이 아닌 실수 전체에 적용 가능한 Yeo-Johnson 변환 적용

```python
# 2016, 2017 금액 변수 데이터셋 생성
all_num_bc = pd.concat([all_imputed.filter(regex='1$'), all_imputed.filter(regex='2$')], axis=1)
```

```python
# 음수 값에도 적용할 수 있도록 Yeo-Johnson 변환
from sklearn.preprocessing import PowerTransformer

all_num_pt = pd.DataFrame()
for col in all_num_bc.columns:
    pt = PowerTransformer(method='yeo-johnson')
    all_num_pt[col] = pt.fit_transform(np.array(all_num_bc[col]).reshape(-1,1))[:,0]

all_num_pt
```

```python
# Yeo-Johnson 변환한 변수 포함한 전체 데이터 생성
all_pt = pd.concat([all_onehot['bedCount'], all_num_pt, all_onehot['runYear'],
                    all_onehot.filter(like='OC'), all_onehot.filter(like='sido'),
                    all_onehot.filter(like='instkind'), all_onehot.filter(like='ownerChange')], axis=1)
all_pt
```

### Log Transformation **
- 2016, 2017 금액 변수 로그 변환 적용
- 수치형 변수 중 bedCount, runYear 변수는 로그 변환 불필요 

```python
# 2016, 2017 금액 변수 데이터셋 생성
all_num = pd.concat([all_onehot.filter(regex='1$'), all_onehot.filter(regex='2$')], axis=1)
```

```python
# 음수 값 포함하는 데이터셋과 포함X 데이터셋 생성
pos_fts = all_num.drop(columns=['profit1','netAsset1','surplus1','profit2','netAsset2','surplus2'], axis=1)
neg_fts = all_num[['profit1','netAsset1','surplus1','profit2','netAsset2','surplus2']] # 음수 값 포함 변수
```

```python
# 음수 값 포함X 변수 skewness 확인
from scipy.stats import skew

skew_features_pos = pos_fts.apply(lambda x : skew(x.dropna()))
print(skew_features_pos.sort_values(ascending=False))
```

```python
# 로그 변환
pos_fts_log = np.log1p(pos_fts)
```

```python
# 음수 값 포함하는 변수 skewness 확인
skew_features_neg = neg_fts.apply(lambda x : skew(x.dropna()))
print(skew_features_neg.sort_values(ascending=False))
```

```python
# 로그 변환
neg_fts_log = np.log1p(neg_fts + abs(neg_fts.min()))
```

### SMOTE
- 임의의 소수 클래스 데이터로부터 인근 소수 클래스 사이에 새로운 데이터를 생성하는 것
- 임의의 소수 클래스에 해당하는 관측치 X를 잡고, 그 X로부터 가장 가까운 K개의 이웃 X를 찾는다. K개의 이웃과 X 사이에 임의의 새로운 데이터 X`를 생성하는 것

```python
from imblearn.over_sampling import SMOTE
from collections import Counter

smote = SMOTE()
X_train_over, y_train_over = smote.fit_resample(X_train_new, y_train_new)

print('SMOTE 적용 전 학습용 피처/레이블 데이터 세트: ', X_train.shape, y_train.shape)
print('SMOTE 적용 후 학습용 피처/레이블 데이터 세트: ', X_train_over.shape, y_train_over.shape)
print("")
print('Resampled dataset shape %s' %Counter(y_train))
print('Resampled dataset shape %s' %Counter(y_train_over))
```

### Borderline SMOTE
- SMOTE에서 조금 변주를 준 알고리즘
- 다수 클래스와 소수 클래스가 서로 인접해 있는 경계선에 있는 소수 클래스의 데이터에 대해서 SMOTE를 적용한 것

```python
from imblearn.over_sampling import BorderlineSMOTE

bsmote = BorderlineSMOTE()
X_btrain_over, y_btrain_over = bsmote.fit_resample(X_train_new, y_train_new)

print('SMOTE 적용 전 학습용 피처/레이블 데이터 세트: ', X_train_new.shape, y_train_new.shape)
print('SMOTE 적용 후 학습용 피처/레이블 데이터 세트: ', X_train_over.shape, y_train_over.shape)
print("")
print('Resampled dataset shape %s' %Counter(y_train_new))
print('Resampled dataset shape %s' %Counter(y_train_over))
```

### ADASYN (Adaptive Synthetic Sampling)
- BorderlineSMOTE에서 조금 더 변주를 준 알고리즘
- 가중치를 통해 SMOTE를 적용시키는 방식

```python
from imblearn.over_sampling import ADASYN
ads = ADASYN()
X_adstrain_over, y_adstrain_over = ads.fit_resample(X_train_new, y_train_new)

print('SMOTE 적용 전 학습용 피처/레이블 데이터 세트: ', X_train_new.shape, y_train_new.shape)
print('SMOTE 적용 후 학습용 피처/레이블 데이터 세트: ', X_adstrain_over.shape, y_adstrain_over.shape)
print("")
print('Resampled dataset shape %s' %Counter(y_train_new))
print('Resampled dataset shape %s' %Counter(y_train_over))
```

```python
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, GradientBoostingClassifier, ExtraTreesClassifier, VotingClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.svm import SVC
from xgboost import XGBClassifier
from sklearn.model_selection import GridSearchCV, cross_val_score, StratifiedKFold, learning_curve
```

```python
kfold = StratifiedKFold(n_splits=10)

random_state = 2
classifiers = []
classifiers.append(SVC(random_state=random_state))
classifiers.append(DecisionTreeClassifier(random_state=random_state))
classifiers.append(AdaBoostClassifier(DecisionTreeClassifier(random_state=random_state),random_state=random_state,learning_rate=0.1))
classifiers.append(RandomForestClassifier(random_state=random_state))
classifiers.append(ExtraTreesClassifier(random_state=random_state))
classifiers.append(GradientBoostingClassifier(random_state=random_state))
classifiers.append(MLPClassifier(random_state=random_state))
classifiers.append(KNeighborsClassifier())
classifiers.append(LogisticRegression(random_state = random_state))
classifiers.append(LinearDiscriminantAnalysis())
classifiers.append(XGBClassifier())

cv_results = []
for classifier in classifiers:
  cv_results.append(cross_val_score(classifier,X_train_over,y_train_over,scoring='accuracy',cv=kfold,n_jobs=4))

cv_means = []
cv_std = []
for cv_result in cv_results:
    cv_means.append(cv_result.mean())
    cv_std.append(cv_result.std())

cv_res = pd.DataFrame({"CrossValMeans":cv_means,"CrossValerrors": cv_std,
                       "Algorithm":["SVC","DecisionTree","AdaBoost","RandomForest","ExtraTrees","GradientBoosting","MultipleLayerPerceptron",
                                    "KNeighboors","LogisticRegression","LinearDiscriminantAnalysis", "XGB"]})
```
